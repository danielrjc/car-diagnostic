```python
import os
import matplotlib.pyplot as plt

#for loading and visualizing audio files
import librosa
import librosa.display

#to play audio
import IPython.display as ipd

audio_fpath = "/content/drive/MyDrive/CPSC 291 team dino/Dataset/Combined Dataset/combined audio/healthy/"
audio_clips = os.listdir(audio_fpath)
print("No. of .mp3 files in audio folder = ",len(audio_clips))
```

```python
x, sr = librosa.load(audio_fpath+audio_clips[0], sr=44100)

print(type(x), type(sr))
print(x.shape, sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

Log transform is done because human hearing works in logarithmic scale

```python
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
plt.colorbar()
save_name = "1.png"
plt.savefig(save_name, quality=100, bbox_inches=0, pad_inches=0)
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
fig = plt.figure(figsize=(5, 5), dpi=1000, frameon=False)
ax = fig.add_axes([0, 0, 1, 1], frameon=False)
ax.axis('off')
librosa.display.specshow(Xdb, sr=sr, cmap='gray', x_axis='time', y_axis='log')
save_name = "2.png"
plt.savefig(save_name, quality=100, bbox_inches=0, pad_inches=0)
#plt.show()
```
