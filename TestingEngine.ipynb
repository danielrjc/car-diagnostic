# Creating the dataframe using pandas

# Downloading the audio files

```python
for i in df_csv['Video']:
  yt = YouTube(i)
  yt.streams.filter(file_extension='mp4', only_audio=True).first().download(output_path='dataset/',timeout=5, skip_existing=True)
  time.sleep(5)
```

# Clipping the files to 10 seconds 

```python
from pytube.helpers import safe_filename
```

```python
from moviepy.video.io.ffmpeg_tools import ffmpeg_extract_subclip


x = 0
for i in df_csv['Video']:
  yt = YouTube(i)
  name = yt.title
  time_string = df_csv['Start Time'][x]
  x += 1
  date_time = datetime.datetime.strptime(time_string, "%M:%S")
  a_timedelta = date_time - datetime.datetime(1900, 1, 1)
  seconds = a_timedelta.total_seconds()

  print(seconds)
  ffmpeg_extract_subclip("dataset/"+safe_filename(name)+".mp4", seconds,seconds+10, targetname="result/"+safe_filename(name)+".mp4")
```
